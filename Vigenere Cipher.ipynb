```python
# Directs you to either the encrypt or decrypt function
def get_input():
    answer = input('Do you want to encrypt or decrypt a message?\n')
    answer = answer.strip().replace('.', '').lower()

    while answer != 'encrypt' and answer != 'decrypt':
        answer = input('Please respond with encrypt or decrypt:\n')
        answer = answer.strip().lower()
        if answer == 'encrypt' or answer == 'decrypt':
            break
        
    if answer == 'encrypt':
        print(vigenere_encrypter())
    
    else:
        print(vigenere_decrypter())

#Checks to see if program should repeat or quit        
def proceed():
    proceed = input('Would you like to encrypt or decrypt another message? Please reply with \'yes\' or \'no\'.\n')
    proceed = proceed.strip().lower()
    
    while proceed != 'yes' and proceed != 'no':
        proceed = input('Please respond with \'yes\' or \'no\'.')
        proceed = proceed.strip().lower()
        if proceed == 'yes':
            return True
            break
        elif proceed == 'no':
            return False
            break 

# This is the function to encrypt messages
def vigenere_encrypter():
    message_to_encrypt = input('Enter the message you wish to encrypt:\n')
    message_to_encrypt = message_to_encrypt.strip()
    
    keyword = input('Enter the keyword you wish to use to encrypt your message:\n')
    keyword = keyword.strip().lower()  
    
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    keyword_phrase = ""
    iterator = 0
    encrypted_message = ""
    
    for character in keyword:
        while character not in alphabet:
            keyword = input('Please enter a keyword that only contains letters:\n')
            keyword = keyword.strip().lower()  
            for character in keyword:
                if character in alphabet:
                    continue
                else:
                    break  
            
    for character in message_to_encrypt:
        if character in alphabet:
            keyword_phrase += keyword[iterator]
            if iterator >= len(keyword) - 1:
                iterator = 0
            else:
                iterator += 1
                
        else:
            keyword_phrase += character
    
    iterator = 0
            
    for character in message_to_encrypt:
        if character in alphabet:
            encrypted_index = alphabet.index(character) + alphabet.index(keyword_phrase[iterator])
            encrypted_message += alphabet[encrypted_index % 26]
            iterator += 1
        else:
            encrypted_message += character
            iterator += 1
            
    return encrypted_message
    

#This is the function to decrypt messages
def vigenere_decrypter():
    message_to_decrypt = input('Enter the message you wish to decrypt:\n')
    message_to_decrypt = message_to_decrypt.strip()
    
    keyword= input('Enter the keyword you wish to use to decrypt your message:\n')
    keyword = keyword.strip().lower()
    
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    keyword_phrase = ""
    iterator = 0
    decrypted_message = ""
    
    for character in keyword:
        while character not in alphabet:
            keyword = input('Please enter a keyword that only contains letters:\n')
            keyword = keyword.strip().lower()
            for character in keyword:
                if character in alphabet:
                    continue
                else:
                    break
    
    for character in message_to_decrypt:
        if character in alphabet:
            keyword_phrase += keyword[iterator]
            if iterator >= len(keyword) - 1:
                iterator = 0
            else:
                iterator += 1
                
        else:
            keyword_phrase += character
    
    iterator = 0
            
    for character in message_to_decrypt:
        if character in alphabet:
            decrypted_index = alphabet.index(character) - alphabet.index(keyword_phrase[iterator])
            decrypted_message += alphabet[decrypted_index]
            iterator += 1
        else:
            decrypted_message += character
            iterator += 1
            
    return decrypted_message

###################
###################
#Execute script below
continue_script = True


while continue_script == True:
    get_input()
    continue_script = proceed()
```
